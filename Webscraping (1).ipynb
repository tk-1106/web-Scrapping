```python
import requests
import pandas 
from bs4 import BeautifulSoup
```

## target = https://www.dominos.co.in/store-location/chennai

```python
request = requests.get('https://www.dominos.co.in/store-location/chennai')
request
```

```python
request.content
```

```python
soup = BeautifulSoup(request.content,"html.parser")
soup
```

```python
branch = soup.find_all('p',{"class":"city-main-sub-title"})
branch
```

```python
for i in branch:
    print(i.text)
```

```python
location = [i.text for i in branch]
location
```

```python
# address = soup.find_all("p", {"class":"grey-text mb-0"})
address = soup.find_all("p",{"class":"grey-text mb-0"})
```

```python
for i in address:
    print(i.text)
```

```python
address = [i.text for i in address]
address
```

```python
Timing = soup.find_all("div",{"class":"col-xs-9 col-md-9 pl0 search-grid-right-text"})
Timing
```

```python
for i in Timing:
    print((i.text).strip())
```

```python
timing = [(i.text).strip() for i in Timing]
timing
```

```python
timing = [(i.text).strip("\n\t") for i in Timing]
timing
```

```python
len(timing)
```

```python
len(address)
```

```python
len(location)
```

```python
data = {"Location":location,"Address":address,"Timing":timing}
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```
